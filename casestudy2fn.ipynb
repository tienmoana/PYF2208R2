Quản lý kho, các chức năng thêm, xóa, sửa, reset data

```python

from tkinter import *
from tkinter import messagebox

class stock:
    def __init__(self, code, name, quantity, price):
        self.code = code
        self.name = name
        self.quantity = quantity
        self.price = price

list_stock = []
        
def addstock():
    list_stock.append(stock(code.get(), name.get(), quantity.get(), price.get()))
    lst_stock.insert(END, list_stock[len(list_stock)-1].code + '   ' + list_stock[len(list_stock)-1].name + '   ' + list_stock[len(list_stock)-1].quantity + '   ' + list_stock[len(list_stock)-1].price)

def repaid():
    check = code.get()
    exit = False
    for i in range(len(list_stock)):
        if list_stock[i].code == check:
            if code.get() != '':
                list_stock[i].code = code.get()
            if name.get() != '':
                list_stock[i].name = name.get()
            if quantity.get() != '':
                list_stock[i].quantity = quantity.get() 
            if price.get() != '':
                list_stock[i].price = price.get()
            lst_stock.delete(i)
            lst_stock.insert(i, str(list_stock[i].code) + '   ' + list_stock[i].name + '   ' + list_stock[i].quantity + '   ' + str(list_stock[i].price))
            messagebox.showinfo('Thông báo', f'Giá trị mới là: {list_stock[i].code}, {list_stock[i].name}, {list_stock[i].quantity}, {list_stock[i].price}')
            exit = True
            break
    if exit == False:
        messagebox.showinfo('Thông báo', f'Không tìm thấy sản phẩm {check}')

def deletestock():
    check = code.get()
    exit = False
    for i in range(len(list_stock)):
        if list_stock[i].code == check:
            del list_stock[i]
            lst_stock.delete(i)
            messagebox.showinfo('Thông báo', f'Đã xóa thành công nhân sản phẩm code {check}')
            exit = True
            break
    if exit == False:
        messagebox.showinfo('Thông báo', f'Không tìm thấy sản phẩm {check}')

def reset():
    code.delete(0, len(code.get()))
    name.delete(0, len(name.get()))
    quantity.delete(0, len(quantity.get()))
    price.delete(0, len(price.get()))
        
windown = Tk()
windown.title('STOCK MANAGEMRNT')

title_lbl = Label(windown, text='STOCK MANAGEMENT', bg = 'yellow', font='arial 25 bold', width=20, height=1)
code_lbl = Label(windown, text='CODE', width=8, bg = 'blue', fg='white', font='arial 12 bold')
name_lbl = Label(windown, text='NAME', width=8, bg = 'blue', fg='white', font='arial 12 bold')
quantity_lbl = Label(windown, text='QUANTITY', width=8, bg = 'blue', fg='white', font='arial 12 bold')
price_lbl = Label(windown, text='PRICE', width=8, bg = 'blue', fg='white', font='arial 12 bold')
thekho_lbl = Label(windown, text='THẺ KHO', font='arial 12 bold')
xoa_lbl = Label(windown, text='DELETE', font='arial 12 bold')
thekhochitiet_lbl = Label(windown, text='CHI TIẾT THẺ KHO', font='arial 12 bold')
code = Entry(windown, relief=SUNKEN, width=20, font='arial 12')
name = Entry(windown, relief=SUNKEN, width=20, font='arial 12')
quantity = Entry(windown, relief=SUNKEN, width=20, font='arial 12')
price = Entry(windown, relief=SUNKEN, width=20, font='arial 12')
add_btn = Button(windown, bg='grey', fg='white', activebackground='blue', text='ADD', width=7, font='arial 12 bold', command=addstock)
repaid_btn = Button(windown, bg='grey', fg='white', activebackground='blue', text='REPAID', width=7, font='arial 12 bold', command=repaid)
delete_btn = Button(windown, bg='grey', fg='white', activebackground='blue', text='DELETE', width=7, font='arial 12 bold', command=deletestock)
reset_btn = Button(windown, bg='grey', fg='white', activebackground='blue', text='RESET', width=7, font='arial 12 bold', command=reset)
lst_stock = Listbox(windown, selectmode=MULTIPLE, relief=SUNKEN, width=50, height=8)
lst_quantity = Listbox(windown, selectmode=MULTIPLE, relief=SUNKEN, width=50, height=8)

title_lbl.grid(row=0, column=2)
thekho_lbl.grid(row=1, column=2)
code_lbl.grid(row=2, column=1)
name_lbl.grid(row=3, column=1)
quantity_lbl.grid(row=4, column=1)
price_lbl.grid(row=5, column=1)
code.grid(row=2, column=2)
name.grid(row=3, column=2)
quantity.grid(row=4, column=2)
price.grid(row=5, column=2)
add_btn.grid(row=2, column=3)
repaid_btn.grid(row=3, column=3)
delete_btn.grid(row=4, column=3)
reset_btn.grid(row=5, column=3)
thekhochitiet_lbl.grid(row=8, column=2)
lst_stock.grid(row=9, column=2)

windown.mainloop()
```
